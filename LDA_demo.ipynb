```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

print(X.shape)
print(len(np.unique(y)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=20)
```

```python
n_components=min(X.shape[1], len(np.unique(y))-1)
print(n_components)

lda = LinearDiscriminantAnalysis(solver='svd', n_components=n_components)
lda.fit(X_train, y_train)
```

```python
lda.score(X_test, y_test)
```

```python
X_r2 = lda.transform(X_train)
```

```python
plt.rcParams['font.size'] = 14
plt.figure(figsize=(10, 8))
colors = ['navy', 'turquoise', 'darkorange']
for color, i, target_name in zip(colors, np.unique(y), target_names):
    plt.scatter(X_r2[y_train == i, 0], X_r2[y_train == i, 1], alpha=.8, color=color,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('LDA of IRIS dataset')
plt.show()
```
